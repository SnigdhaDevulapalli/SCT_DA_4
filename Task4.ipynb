```python
numeric_cols = ['Impressions','Clicks','Leads','Conversions','Revenue','Acquisition_Cost','ROI']
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(df[numeric_cols].corr(),annot=True,cmap='coolwarm'
plt.title("Correlation Matrix")
plt.show()
```

```python
impressions = df['Impressions'].sum()
clicks = df['Clicks'].sum()
leads = df['Leads'].sum()
conversions = df['Conversions'].sum()
```

```python
funnel = pd.DataFrame({'Stage':['Impressions','Clicks','Leads','Conversions'],'Count':[impressions,clicks,leads,conversions]})
print(funnel)
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Stage',y='Count',data=funnel)
plt.title("Marketing Funnel")
plt.show()
```

```python
channel_roi = df.groupby('Channel_Used')['ROI'].mean()
```

```python
channel_roi = channel_roi.sort_values(ascending=False)
print(channel_roi)
```

```python
channels = df['Channel_Used'].str.split(', ')
```

```python
all_channels = channels.explode()
```

```python
channel_counts = all_channels.value_counts()
print(channel_counts)
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=channel_counts.index,y=channel_counts.values)
plt.title('Marketing Channels')
plt.xlabel('Channel')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
channel_revenue = df.groupby('Channel_Used')['Revenue'].mean()
```

```python
channel_revenue = df.groupby('Channel_Used')['Revenue'].mean()
top_10 = channel_revenue.sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
top_10.plot(kind='barh')
plt.xlabel('Average Revenue')
plt.ylabel('Channel Used')
plt.title('Top 10 Channels by Average Revenue')
plt.show()
```

# Marketing Campaign Analysis Report

## Introduction
This analysis was performed on the Nykaa marketing campaign dataset containing 55,555 records. Data cleaning showed that there were no missing values or duplicate records, indicating high data quality.

## Exploratory Data Analysis
Several visualizations were created to understand campaign performance. The marketing funnel showed how customers progressed from impressions to clicks, leads, and conversions. A decline was observed at each stage, indicating opportunities to improve conversion rates.

Correlation analysis revealed positive relationships among impressions, clicks, leads, conversions, and revenue. Revenue increased as customer engagement increased.

## ROI Analysis
Average ROI was calculated for each marketing channel. Some channels generated significantly higher returns while maintaining lower acquisition costs. These channels were found to be more efficient in generating revenue.

## Recommendations
Based on ROI and revenue analysis, more marketing budget should be allocated to the top-performing channels. Channels with lower ROI should be optimized or receive reduced spending. Focusing investment on the best-performing channels can improve overall campaign efficiency and maximize returns.

## Conclusion
Exploratory Data Analysis provided valuable insights into customer behavior and channel effectiveness. Strategic allocation of marketing budget toward high-ROI channels can enhance campaign performance and increase profitability.
